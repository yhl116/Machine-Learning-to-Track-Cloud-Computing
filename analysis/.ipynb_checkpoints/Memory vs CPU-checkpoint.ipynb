```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
from random import randrange
import progressbar

SELECT_NUMBER = 1000

random.seed(1)
x_axis = np.arange(0, 29-1/288, 1/288).tolist()
np_cpu = np.load("wetransfer-2ee104//google-cpu-full.npy")
np_memory = np.load("wetransfer-2ee104//google-mem-full.npy")

random_samples = [34, 150, 352, 365, 416, 464, 475, 501, 565, 1033, 1416, 1537, 1638, 1674, 1768, 1931, 1980, 2201, 
                  2682, 2834, 3045, 3110, 3439, 3548, 3584, 3632, 3748, 3782, 3818, 3977, 4179, 4363, 4655, 4747, 
                  4856, 4970, 5200, 5450, 5663, 6014, 6070, 6139, 6219, 6245, 6386, 6443, 6444, 6623, 6788, 6818, 
                  6915, 7090, 7174, 7191, 7297, 7364, 7530, 7737, 7868, 7993, 8023, 8117, 8123, 8181, 8205, 8278, 
                  8318, 8330, 8535, 8644, 8870, 8991, 9058, 9116, 9325, 9626, 9650, 9685, 9952, 10311, 10523, 10642,
                  10675, 10875, 10891, 10982, 11050, 11089, 11247, 11400, 11518, 11652, 11820, 11821, 11856, 11892,
                  12094, 12175, 12185, 12467]  

df_cpu = pd.DataFrame(np_cpu).transpose()
df_memory = pd.DataFrame(np_memory).transpose()
```

```python
# dropping the last 32 rows as they are all zeros and give us no information

df_cpu = df_cpu.iloc[:-33]
df_memory = df_memory.iloc[:-33]
```

```python
# get the correlation between cpu and memory for each machine
mem_cpu_correlation = list()

with progressbar.ProgressBar(max_value = len(df_cpu.columns)) as bar:
    for x in range (0,len(df_cpu.columns)):
        
        bar.update(x)
        temp_df = pd.DataFrame({"mem": np_memory[x][:-33], "cpu": np_cpu[x][:-33]})
        mem_cpu_correlation.append(temp_df.corr().iloc[1][0])
        
        
```

```python
# plotting the distribution of the maximum correlation

plt.figure(figsize=(12,6))
binning = np.arange(-1, 1.05, 0.02).tolist()
value_of_each_bin = plt.hist(mem_cpu_correlation, bins = binning)
plt.minorticks_on()
plt.rc('font', **font)
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title("CPU-Memory Correlation")
plt.xlabel("CPU-Memory Correlation")
plt.ylabel("Number of Occurences")
plt.savefig("figures//analysis//CPU-Memory Correlation")
```

```python
sorted_mem_cpu_correlation = sorted(mem_cpu_correlation)

print("Mean of CPU-Memory Correlation: ", np.mean(mem_cpu_correlation))
print("CPU-Memory Correlation of last 1000 machines > ", sorted_mem_cpu_correlation[-1000])
print(465/len(np_cpu)*100, "% of machines have cpu-memory correlation of > 0.7")
```

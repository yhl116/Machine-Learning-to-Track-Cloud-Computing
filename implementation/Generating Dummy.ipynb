```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
from random import randrange
import progressbar
import math
```

```python
x_axis = np.arange(0,6)
y1 = list()
y2 = list()
y3 = list()
y4 = list()

for x in x_axis:
    y1.append(15*math.sin(x*0.4+1)+2)
    y2.append(20*math.sin(x*0.4+1)+15)
    y3.append(2*x+25)
    y4.append(-1*x+15)
```

```python
plt.figure(figsize = (9,6))

plt.title("Dummy Examples",fontsize = 18)

plt.plot(y1[0:5], label = "Example Machine 1")
plt.plot(y2[0:5], label = "Example Machine 2")
plt.plot(x_axis[0:5], y3[0:5], label = "Example Machine 3")
plt.plot(x_axis[0:5], y4[0:5], label = "Example Machine 4")

plt.scatter(x_axis[0:5], y1[0:5])
plt.scatter(x_axis[0:5], y2[0:5])
plt.scatter(x_axis[0:5], y3[0:5])
plt.scatter(x_axis[0:5], y4[0:5])

plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()
plt.savefig("figures//proposed//dummy")
```

```python
plt.figure(figsize = (9,6))

plt.title("Trajectory of Dummy Examples",fontsize = 18)

plt.plot(y1[0:5], label = "Example Machine 1")
plt.plot(y2[0:5], label = "Example Machine 2")
plt.plot(x_axis[0:5], y3[0:5], label = "Example Machine 3")
plt.plot(x_axis[0:5], y4[0:5], label = "Example Machine 4")

plt.scatter(x_axis[0:5], y1[0:5])
plt.scatter(x_axis[0:5], y2[0:5])
plt.scatter(x_axis[0:5], y3[0:5])
plt.scatter(x_axis[0:5], y4[0:5])

plt.scatter(x_axis[5], y1[5], marker = 'x', color='#1f77b4')
plt.scatter(x_axis[5], y2[5], marker = 'x', color='#ff7f0e')
plt.scatter(x_axis[5], y3[5], marker = 'x', color='#2ca02c')
plt.scatter(x_axis[5], y4[5], marker = 'x', color='#d62728')

plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()
plt.savefig("figures//proposed//trajectory")
```

```python
plt.figure(figsize = (16,5))

plt.subplot(1,2,1)
plt.plot(y1[0:5], label = "Example Machine 1")
plt.plot(y2[0:5], label = "Example Machine 2")
plt.scatter(x_axis[0:5], y1[0:5])
plt.scatter(x_axis[0:5], y2[0:5])

plt.title("Cluster A",fontsize = 18)
plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()


plt.subplot(1,2,2)
plt.plot(x_axis[0:5], y3[0:5], label = "Example Machine 3", color='#2ca02c')
plt.plot(x_axis[0:5], y4[0:5], label = "Example Machine 4", color='#d62728')
plt.scatter(x_axis[0:5], y3[0:5], color='#2ca02c')
plt.scatter(x_axis[0:5], y4[0:5], color='#d62728')

plt.title("Cluster B",fontsize = 18)
plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()


plt.savefig("figures//proposed//cluster")
```

```python
y5 = list()
y6 = list()

for x in x_axis:
    y5.append(17.5*math.sin(x*0.4+1)+8.5)
    y6.append(0.5*x+20)

plt.figure(figsize = (16,5))

plt.subplot(1,2,1)
plt.plot(y1[0:5], label = "Example Machine 1")
plt.plot(y2[0:5], label = "Example Machine 2")
plt.scatter(x_axis[0:5], y1[0:5])
plt.scatter(x_axis[0:5], y2[0:5])

plt.plot(y5[0:5], label = "General Model A", color='#9467bd')
plt.scatter(x_axis[0:5], y5[0:5], color='#9467bd')
plt.scatter(x_axis[5], y5[5], color='#9467bd', marker = "x")

plt.title("Cluster A",fontsize = 18)
plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()


plt.subplot(1,2,2)
plt.plot(x_axis[0:5], y3[0:5], label = "Example Machine 3", color='#2ca02c')
plt.plot(x_axis[0:5], y4[0:5], label = "Example Machine 4", color='#d62728')
plt.scatter(x_axis[0:5], y3[0:5], color='#2ca02c')
plt.scatter(x_axis[0:5], y4[0:5], color='#d62728')

plt.plot(y6[0:5], label = "General Model B", color='#8c564b')
plt.scatter(x_axis[0:5], y6[0:5], color='#8c564b')
plt.scatter(x_axis[5], y6[5], color='#8c564b', marker = "x")

plt.title("Cluster 2B",fontsize = 18)
plt.ylim(0,40)
plt.xlim(-0.25, 5.25)
plt.legend(loc = 3)
plt.xticks(np.arange(0,6), ["n-4", "n-3", "n-2", "n-1", "n", "n+1"])
plt.xlabel("Sample", fontsize = 16)
plt.ylabel("Mean CPU Usage", fontsize = 16)
plt.grid()


plt.savefig("figures//proposed//general_model")
```

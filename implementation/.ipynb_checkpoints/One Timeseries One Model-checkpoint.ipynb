```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import progressbar
from statsmodels.tsa.arima_model import ARIMA

SELECT_NUMBER = 1000

x_axis = np.arange(0, 29-1/288, 1/288).tolist()
np_cpu = np.load("dataset//google-cpu-full.npy")

# full_cpu is a list of pandas series of the cpu utilisation of all machines
# just a lazy way of converting nd-arrays to pd-series
full_cpu = []
for x in range (0,len(np_cpu)):
    full_cpu.append(pd.Series(np_cpu[x], index = x_axis))  
    
random_samples = [34, 150, 352, 365, 416, 464, 475, 501, 565, 1033, 1416, 1537, 1638, 1674, 1768, 1931, 1980, 2201, 
                  2682, 2834, 3045, 3110, 3439, 3548, 3584, 3632, 3748, 3782, 3818, 3977, 4179, 4363, 4655, 4747, 
                  4856, 4970, 5200, 5450, 5663, 6014, 6070, 6139, 6219, 6245, 6386, 6443, 6444, 6623, 6788, 6818, 
                  6915, 7090, 7174, 7191, 7297, 7364, 7530, 7737, 7868, 7993, 8023, 8117, 8123, 8181, 8205, 8278, 
                  8318, 8330, 8535, 8644, 8870, 8991, 9058, 9116, 9325, 9626, 9650, 9685, 9952, 10311, 10523, 10642,
                  10675, 10875, 10891, 10982, 11050, 11089, 11247, 11400, 11518, 11652, 11820, 11821, 11856, 11892,
                  12094, 12175, 12185, 12467]    
```

```python
selected_machines = pd.read_csv("selected_machines_ls.csv", header = None)
selected_machines = selected_machines[1]

# df_cpu is a dataframe with all the cpu utilisation data (columns = machine; rows = time)
# dropping last 32 lines with corrupted data

df_cpu = pd.DataFrame(np_cpu).transpose()
df_selected_machines = df_cpu[selected_machines].drop(df_cpu.tail(32).index)
```

```python
def get_machine_predictions(machine_timeseries, start_time, end_time, input_arima_order = (3,0,0)):
    # takes in timeseries of one machine
    # machine_timeseries is a ps.series
    # returns a list of predicted values of time start_time+1 to end_time+1
    
    machine_prediction = []
    
    for current_time in range (start_time, end_time):
        # when current_time = t, generates predictions for t+1
        
        current_timeseries = machine_timeseries[:current_time].values
        model = ARIMA(current_timeseries, order = input_arima_order)
        model_fit = model.fit(disp=0)
        current_prediction = model_fit.forecast()[0][0]
        
        machine_prediction.append(current_prediction)
    
    return machine_prediction

def get_all_predictions(cpu_data, start_time, end_time = None):
    arima_order = (3,0,0)
    machine_predictions_map = dict()
    
    if end_time == None:
        end_time = len(cpu_data.index) - 1
        
    with progressbar.ProgressBar(max_value = len(cpu_data.columns)) as bar:
        
        # iterate through all machines 
        for machine_count,current_machine in enumerate(cpu_data):
            
            bar.update(machine_count)    
            
            # current_machine_timeseries is the timeseries of current_machine
            # current_machine is the index of the required machine
            current_machine_timeseries = cpu_data[current_machine]
            
            # get_arima_predictions returns all the predictions of current machine throughout all required time index
            current_machine_predictions = get_machine_predictions(current_machine_timeseries, start_time, end_time, input_arima_order = arima_order)
            
            # add predictions to map
            machine_predictions_map[current_machine] = current_machine_predictions
            
    time_index_of_prediction = np.arange(start_time, end_time)        
            
    return pd.DataFrame(machine_predictions_map, index = time_index_of_prediction)
    
```

```python
test_predictions = get_all_predictions(df_selected_machines, start_time = 288, end_time = 288+5)
```

```python
mse = (test_predictions - df_selected_machines.iloc[289:361])**2
mse.mean().mean()
```

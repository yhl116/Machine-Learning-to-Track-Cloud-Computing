```python
list0 = [1,2,5,6,4,5,-1,4]

def smallest(list1):
    list1 = list(filter(lambda x: x > 0, list1))

    if(len(list1) == 0):
        return 0
    else:
        list1 = sorted(list1)
        curr = list1[0]
        for i, x in enumerate(list1):
            if x == curr + 1:
                curr = x
            elif x > curr + 1:
                return curr + 1
            
        return list1[-1] + 1
```

```python
def parse_minute(minute_string):
    '''minute_sting -- 

        return -- a list of minutes instances specified by cron_string
    '''

    minute_output = []
    minute_string = minute_string.replace(" ","")
    minute_list = minute_string.split(",")

    for current_string in minute_list:
        if(current_string.find("-") != -1):
            try:
                minute_range = current_string.split("-")
                if(len(minute_range) != 2):
                    raise Exception("Minute range must be in the form a-b where a and b are inegers between 0-59: " + current_string)

                start_range = int(minute_range[0])
                end_range = int(minute_range[1])
                if not(start_range >= 0 and start_range <= 59 and end_range >= 0 and end_range <= 59 and end_range > start_range):
                    raise Exception("Minute out of range, must be between 0-59: " + current_string)

                for x in range (start_range, end_range + 1):
                    minute_output.append(x)

            except Exception as e:
                print(e)     
                raise Exception("Invalid minute range: " + current_string)   

        elif(current_string.find("/") != -1):
            try:
                minute_range = current_string.split("/")
                divider = int(minute_range[1])
                if(minute_range[0] == "*"):
                    start_minute = 0
                else:
                    start_minute = int(minute_range[0])
                

                if(len(minute_range) != 2):
                    raise Exception("Minute range must be in the form a/b where a is the initial minute and b is the subsequent minutes: " + current_string)

                if not(divider >= 0 and divider <= 59 and start_minute >= 0 and start_minute <= 59):
                    raise Exception("Minute out of range, must be between 0-59: " + current_string)  

                for x in range(start_minute, 60, divider):
                    minute_output.append(x)

            except Exception as e:
                print(e)
                raise Exception("Invalid minute input: " + current_string)     

        else:
            try:
                if(current_string == "*"):
                    for x in range (0,60):
                        minute_output.append(x)
                else:
                    int_minute = int(current_string)
                    if(int_minute >= 0 and int_minute <= 59):
                        minute_output.append(int_minute)
                    else:
                        raise Exception("Input minute not in valid range: " + str(int_minute))
            except Exception as e:
                print(e)
                raise Exception("Invalid minute input")

    minute_output = list(dict.fromkeys(minute_output))
    return minute_output            
                
```

```python
def parse_hour(hour_string):
    '''hour_sting -- 

        return -- a list of hours instances specified by cron_string
    '''

    hour_output = []
    hour_string = hour_string.replace(" ","")
    hour_list = hour_string.split(",")

    for current_string in hour_list:
        if(current_string.find("-") != -1):
            try:
                hour_range = current_string.split("-")
                if(len(hour_range) != 2):
                    raise Exception("hour range must be in the form a-b where a and b are inegers between 0-23: " + current_string)

                start_range = int(hour_range[0])
                end_range = int(hour_range[1])
                if not(start_range >= 0 and start_range <= 23 and end_range >= 0 and end_range <= 23 and end_range > start_range):
                    raise Exception("hour out of range, must be between 0-23: " + current_string)

                for x in range (start_range, end_range + 1):
                    hour_output.append(x)

            except Exception as e:
                print(e)     
                raise Exception("Invalid hour range: " + current_string)   

        elif(current_string.find("/") != -1):
            try:
                hour_range = current_string.split("/")
                divider = int(hour_range[1])
                if(hour_range[0] == "*"):
                    start_hour = 0
                else:
                    start_hour = int(hour_range[0])
                

                if(len(hour_range) != 2):
                    raise Exception("hour range must be in the form a/b where a is the initial hour and b is the subsequent hours: " + current_string)

                if not(divider >= 0 and divider <= 23 and start_hour >= 0 and start_hour <= 23):
                    raise Exception("hour out of range, must be between 0-23: " + current_string)  

                for x in range(start_hour, 24, divider):
                    hour_output.append(x)

            except Exception as e:
                print(e)
                raise Exception("Invalid hour input: " + current_string)     

        else:
            try:
                if(current_string == "*"):
                    for x in range (0,24):
                        hour_output.append(x)
                else:
                    int_hour = int(current_string)
                    if(int_hour >= 0 and int_hour <= 23):
                        hour_output.append(int_hour)
                    else:
                        raise Exception("Input hour not in valid range: " + str(int_hour))
            except Exception as e:
                print(e)
                raise Exception("Invalid hour input")

    hour_output = list(dict.fromkeys(hour_output))
    hour_output.sort()
    if(len(hour_output) > 14):
        hour_output = hour_output[0:14]

    return hour_output
```
